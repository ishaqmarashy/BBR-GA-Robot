```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'BBR_sensor.csv'  
data = pd.read_csv(file_path, header=None, index_col=0)

data.index = pd.to_datetime(data.index, unit='ms')

data_resampled = data.resample('1S').mean()

data_resampled.index = (data_resampled.index - data_resampled.index.min()).total_seconds()

plt.plot(data_resampled.index, data_resampled.iloc[:, 0:3])

plt.legend(['gs0','gs1','gs2'])

plt.xlabel('Time (seconds)')
plt.ylabel('Normalized Mean Values (0-1)')
plt.title('Ground Sensors')

plt.show()
```
