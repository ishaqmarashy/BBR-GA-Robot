```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "./controllers/supervisorGA - ER - T2/fitness_data.csv"
df = pd.read_csv(file_path, header=None, names=['Generations', 'Fitness'])

max_fitness_per_generation = df.groupby('Generations')['Fitness'].max().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(max_fitness_per_generation['Generations'], max_fitness_per_generation['Fitness'], marker='o', linestyle='-', color='b')
plt.title('Highest Fitness vs Generations')
plt.xlabel('Generations')
plt.ylabel('Highest Fitness')
plt.grid(True)
plt.show()
```
